```python
import warnings
warnings.filterwarnings("ignore")
import nltk
import string
nltk.download("punkt")
nltk.download("punkt_tab")
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
text = """Ntural Lnaguage Processing is fun to learn.
It helps computers to understand human language! 
Let's explore tokenization today"""
```

```python
l_text = text.lower()
print( "After lowercasing : \n", l_text)
```

### Sentence Tokenization

```python
sentence = sent_tokenize(l_text)
print("After Sentence Tokenize : \n",sentence, "\n","Total Sentences are" , len(sentence))
```

### Word Tokenization

```python
word = word_tokenize(l_text)
print("After word Tokenize : \n",word, "\n","Total Sentences are" , len(word))
```

```python
for i, s in enumerate(sentence,1):
    print(f"\n Sentence {i}: {s}")
    print("Tokens:", word_tokenize(s))
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
t = ["I love NLP üòç! It‚Äôs amazing to analyze 1000s of messages, emails, and tweets!!!",
    "Cleaning text data isn't easy üòÖ ‚Äî we remove #hashtags, numbers (123), and special chars like @ or $."]
```

```python
len(t)
```

```python
clean_text = []
for i in t:
    clean_text.append(i.lower())
print(clean_text)
```

```python
# removing emojis and specioal characters, and number
def clean_special(text):
    text = re.sub(r'[^a-zA-Z\s]',"",text)
    return text
```

```python
cleaned_text = [clean_special(t) for t in clean_text ]
```

```python
cleaned_text
```

```python
# Tokenize
tokenized_text = []
for i in cleaned_text:
    tokenized_text.append(word_tokenize(i))
    
```

```python
stop_words = set(stopwords.words("english"))
new_text = []
for i in tokenized_text:
    for x in i:
        if x not in stop_words:
            new_text.append(x)
    # print(i)
```

```python
# Stemming
stemmer = PorterStemmer()
for i in new_text:
    print(i,"<---------------->",stemmer.stem(i))
```

```python
# lemmatize
lem = WordNetLemmatizer()
for i in new_text:
    print(i,"<------------------>",lem.lemmatize(i))
```
